## Add sum of constituent params as a column

```python
train_df['sum_charge'] = train_df['constituents_charge'].apply(sum)
```

```python
train_df['sum_Eem'] = train_df['constituents_Eem'].apply(sum)
train_df['sum_Ehad'] = train_df['constituents_Ehad'].apply(sum)
```

```python
train_df.head()
```

## Figure out indices of N hardest constituents

```python
Nhardest = 5
```

```python
xtmp = train_df['constituents_pt'].apply(lambda a: np.argsort(a, kind='heapsort')[-Nhardest:])
```

```python
xtmp.name = 'constituents_pt_idx'
#columns to be unwraped for the Nhardest constituents
columns_arrays = ['constituents_pt', 'constituents_eta',
       'constituents_phi', 'constituents_charge', 'constituents_dxy',
       'constituents_dz', 'constituents_Eem', 'constituents_Ehad']
```

```python
xtmp = pd.concat([xtmp, train_df[columns_arrays]], axis=1)
```

```python
xtmp.columns
```

For each of the array columns extract the parameters for the Nhard constituents

```python
for s in columns_arrays:
    print(s)
    x = xtmp.apply(lambda a: pd.Series(a[s][a['constituents_pt_idx']]), 
                           axis=1)
    print(x.columns)
    new_cols = []
    for i in x.columns:
        new_cols.append('{}_{}'.format(s,i))
    x.columns = new_cols
    train_df = pd.concat([train_df, x], axis=1)
    print(train_df.columns)
```

```python
train_df.memory_usage(index=True, deep=True)
```

### Save the processed sample into pickle (smaller than csv)

```python
(train_df.drop(columns_arrays, axis=1)).to_pickle('test_full_Nhardest{}.pickle'.format(Nhardest))
```
